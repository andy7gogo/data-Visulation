```python
da = df.query("yearID== [2011,2012,2013,2014,2015]").sort_values(by="yearID")

team1 = da["teamID"].values
yea = da["yearID"].values
ERA = da["ERA"].values

s = set(team1)
team = list(s)

year = [2011,2012,2013,2014,2015]
listERA = []
for i in range(len(team)):
    tempall = []
    for j in range(len(year)):
        mean = []
        for k in range(len(team1)):
            if(team[i]==team1[k] and year[j]==yea[k]):
                mean.append(ERA[k])
        
        if(len(mean)!=0):
            sum =0
            for item in mean:
                sum += item
            
            me = sum/len(mean)
            tempall.append(me)
        
        elif(len(mean)==0):
            tempall.append(0)
    
    listERA.append(tempall)

list2011 = []
for i in range(len(listERA)):
    list2011.append(listERA[i][0])
    
list2012 = []
for i in range(len(listERA)):
    list2012.append(listERA[i][1])

list2013 = []
for i in range(len(listERA)):
    list2013.append(listERA[i][2])
    
list2014 = []
for i in range(len(listERA)):
    list2014.append(listERA[i][3])
    
list2015 = []
for i in range(len(listERA)):
    list2015.append(listERA[i][4])
    
    

answer = {"2011":list2011,"2012":list2012,"2013":list2013,"2014":list2014,"2015":list2015}
meanERA = pd.DataFrame(answer,index = team)
meanERA
```
